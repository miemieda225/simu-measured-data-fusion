```python
from Util.read import get_downsample
from Util.read import get_full, cut_data
import os
import shutil
import matplotlib.pyplot as plt
```

```python
def process_folders(input_dir, output_dir):
    # 遍历 test1-17 文件夹
    for i in range(1, 18):
        folder_name = f'test{i}'
        folder_path = os.path.join(input_dir, folder_name)
        result = get_full(folder_path)
        result_cut = cut_data(result)    
        del result
        output_file_path = os.path.join(output_dir, f'{folder_name}.npy')  # 结果保存为文件夹名的文本文件
        with open(output_file_path, 'w') as f:
            f.write(str(result_cut))  # 将结果写入文件
            print(f"Processed {folder_name}, result saved to {output_file_path}")
```

```python
# 设置输入和输出目录
input_dir = r'D:\研三\仿真数据\结构试槽-接缝数据\20241109\data\2024-11-09'  # 你的输入文件夹路径
output_dir = r'D:\研三\仿真数据\结构试槽-接缝数据\20241109\data\2024-11-09\output'   # 你的输出文件夹路径

process_folders(input_dir, output_dir)
```
