```python
import pandas as pd
import numpy as np

# 1. 配置路径和参数
file_path = r'D:\temp\slabshicao\data\Report_P20000.csv'  # 请替换为实际文件名
num_rows = 41          # 纵向提取次数
nodes_per_row = 41     # 每次提取的连续点数
interval = 164         # 提取起点的步长

df = pd.read_csv(file_path, sep=None, engine='python')

    # 3. 清理数据
    # 强制转换 U3 列为数值，无法转换的变为 NaN，然后删除
df[' U3'] = pd.to_numeric(df[' U3'], errors='coerce')
clean_data = df[' U3'].dropna().values

# 4. 矩阵提取
data_matrix = []
for i in range(num_rows):
    start_idx = i * interval
    end_idx = start_idx + nodes_per_row
    print(start_idx, end_idx)
    # 提取切片
    row_values = clean_data[start_idx : end_idx]
    
    # 检查数据长度是否满足要求
    if len(row_values) == nodes_per_row:
        data_matrix.append(row_values)
    else:
        print(f"警告：第 {i} 行数据长度不足，已跳过。")

# 5. 转换为 NumPy 矩阵
result = np.array(data_matrix)

# 6. 保存结果 (可选：保存为 .npy 或 .txt)
np.save('extracted_matrix.npy', result)
# np.savetxt('result.txt', result) # 如果需要文本格式

print("处理完成！矩阵形状为:", result.shape)

```
